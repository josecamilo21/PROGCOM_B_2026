```python
#@title 📈 Simulador de Inversión Eficiente (Quinto Semestre) { run: "auto" }

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- ENTRADA DE DATOS (Interfaz de Usuario) ---
#@markdown Configura cuánto quieres invertir en cada sector (Debe sumar 100%):
tecnologia = 15 #@param {type:"slider", min:0, max:100, step:5}
energia = 5 #@param {type:"slider", min:0, max:100, step:5}
bonos = 15 #@param {type:"slider", min:0, max:100, step:5}
cripto = 65 #@param {type:"slider", min:0, max:100, step:5}

# 1. VALIDACIÓN (Lógica de Programación)
total = tecnologia + energia + bonos + cripto
pesos = np.array([tecnologia, energia, bonos, cripto]) / 100

if total != 100:
    print(f"⚠️ ERROR: Tu inversión suma {total}%. Debe sumar exactamente 100%.")
else:
    # 2. DEFINICIÓN DEL MERCADO (Modelamiento Financiero)
    # Datos: [Retorno Esperado %, Riesgo/Volatilidad %]
    activos = ['Tecnología', 'Energía', 'Bonos', 'Cripto']
    datos_mercado = np.array([
        [18, 25],  # Tecnología: Mucha ganancia, mucho riesgo
        [10, 15],  # Energía: Estable
        [5, 4],    # Bonos: Muy seguro, poca ganancia
        [45, 70]   # Cripto: Extremo
    ])

    # 3. CÁLCULO EFICIENTE (Análisis de Algoritmos & Estadística)
    # Usamos álgebra lineal (vectores) para que sea O(1) en tiempo de ejecución relativo a la entrada
    retornos_esperados = datos_mercado[:, 0]
    riesgos_esperados = datos_mercado[:, 1]

    mi_retorno = np.dot(pesos, retornos_esperados)
    mi_riesgo = np.dot(pesos, riesgos_esperados)

    # 4. VISUALIZACIÓN (Creatividad)
    print(f"✅ PORTAFOLIO VÁLIDO")
    print(f"---")
    print(f"Retorno Anual Esperado: {mi_retorno:.2f}%")
    print(f"Riesgo (Volatilidad): {mi_riesgo:.2f}%")

    # Gráfico de barras para comparar
    plt.figure(figsize=(10, 5))
    colores = ['#4285F4', '#EA4335', '#FBBC05', '#34A853']
    plt.bar(activos, pesos*100, color=colores)
    plt.ylabel('Porcentaje de tu Dinero (%)')
    plt.title('Distribución de tu Capital')
    plt.ylim(0, 100)
    plt.show()
```
