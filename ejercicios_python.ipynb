```python
import math
```

```python
a = float(input("Ingresa el valor de a: "))
b = float(input("Ingresa el valor de b: "))
c = float(input("Ingresa el valor de c: "))
```

```python
discriminante = (b**2-4*a*c)
```

```python
if discriminante > 0:
  x1 = (-b+ math.sqrt(discriminante)) / (2*a)
  x2 = (-b - math.sqrt(discriminante)) / (2*a)
```

```python
if discriminante >= 0:
    x1 = (-b + math.sqrt(discriminante)) / (2*a)
    x2 = (-b - math.sqrt(discriminante)) / (2*a)
    print("Las soluciones son:", x1, "y", x2)
else:
    print("La ecuación no tiene soluciones reales.")
```

Ejercicio 11

```python
nombre = input("Ingresa tu nombre: ")
peso = float(input("Ingresa tu peso en kg: "))
altura = float(input("Ingresa tu altura en metros (ej. 1.75): "))
```

```python
imc = peso / (altura**2)
```

```python
print("\n--- Resultado ---")
print("Usuario:", nombre)
print("Tu IMC es:", round(imc, 2))
```

```python
if imc < 18.5:
    categoria = "Bajo peso"
    desc = "Se recomienda consultar a un nutricionista para aumentar masa muscular."
elif 18.5 <= imc < 24.9:
    categoria = "Normal"
    desc = "¡Felicidades! Mantén tu dieta equilibrada y ejercicio regular."
elif 25 <= imc < 29.9:
    categoria = "Sobrepeso"
    desc = "Considera aumentar tu actividad física y cuidar las porciones."
else:
    categoria = "Obesidad"
    desc = "Es importante acudir al médico para un plan de salud integral."

print("Categoría:", categoria)
print("Recomendación:", desc)
```

Ejercicio 12

```python
n1 = float(input("Primer número: "))
n2 = float(input("Segundo número: "))
operacion = input("Elige operación (+, -, *, /): ")
```

```python
if operacion == "+":
    print("Resultado:", n1 + n2)
elif operacion == "-":
    print("Resultado:", n1 - n2)
elif operacion == "*":
    print("Resultado:", n1 * n2)
elif operacion == "/":
    if n2 != 0:
        print("Resultado:", n1 / n2)
    else:
        print("Error: No se puede dividir por cero.")
else:
    print("Operación no válida.")
```

Ejercicio 15

```python
a = float(input("Lado A: "))
b = float(input("Lado B: "))
c = float(input("Lado C: "))
```

```python
# Primero validamos si existe el triángulo
if (a + b > c) and (a + c > b) and (b + c > a):
    # Si entra aquí, determinamos el tipo
    if a == b == c:
        print("Tipo: Equilátero")
    elif a == b or a == c or b == c:
        print("Tipo: Isósceles")
    else:
        print("Tipo: Escaleno")
else:
    print("Esos lados no pueden formar un triángulo.")
```
