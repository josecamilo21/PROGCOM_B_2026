#Estructuras de selección

##IF/ELSE
Evalua una o mas condiciones, y dependiendo de su valor de la verdad, realizará una acción y otra (sólo dos acciones)

Sintaxis

```
if (condición):
Si la condición es verdadera, se ejecuta este pedazo de código
else:
Si la condición es falsa se ejecuta este pedazo de codigo
```



````python
lluvia=input("Esta lloviendo? ")
````

````python
if lluvia=="si" or lluvia=="si" or lluvia=="si. Muchisimo":
  print("Use sombrilla. Guarde la ropa")
else:
  print("Salga tranquilo")
````

## if ternario
Sintaxis

```
variable= (código si condición es verdadera) if (condición) else (código si condición es falsa)
```



````python
aguacero = "Use sombrilla. Guarde la ropa" if lluvia=="si" else "Salga tranquilo"
````
