Ejercicio 16

```python
num = int(input("¿De qué número quieres la tabla?: "))

print(f"\nTabla del {num}:")
for i in range(1, 11):
    resultado = num * i
    print(f"{num} x {i} = {resultado}")
```

Ejercicio 17

```python
n = int(input("Suma de naturales hasta el número: "))
suma = 0
contador = 1

# Usando ciclo mientras (while)
while contador <= n:
    suma += contador
    contador += 1

print(f"La suma total de 1 hasta {n} es: {suma}")
```

Ejercicio 18

```python
num = int(input("Ingresa un número para calcular su factorial: "))

if num < 0:
    print("No existen factoriales de números negativos.")
elif num == 0:
    print("El factorial de 0 es 1.")
else:
    factorial = 1
    for i in range(1, num + 1):
        factorial *= i
    print(f"El factorial de {num} es {factorial}")
```

Ejercicio 19

```python
n = int(input("¿Cuántos números de la serie Fibonacci quieres ver?: "))

a, b = 0, 1
count = 0

print("Serie Fibonacci:")
while count < n:
    print(a, end=" ")
    # Actualizamos los valores: el nuevo 'a' es 'b', y el nuevo 'b' es la suma
    a, b = b, a + b
    count += 1
```

Ejercicio 20

```python
import random

# 1. Definir lista de 10 colores
colores = ["rojo", "azul", "verde", "amarillo", "negro", "blanco", "rosa", "morado", "naranja", "gris"]
color_secreto = random.choice(colores)
intentos = 10

print("¡Adivina el color secreto!")
print(f"Opciones: {colores}")

while intentos > 0:
    intento_usuario = input(f"\n({intentos} intentos restantes) ¿Qué color crees que es?: ").lower()

    if intento_usuario == color_secreto:
        print("¡Felicidades! Adivinaste el color.")
        break
    else:
        intentos -= 1
        if intentos > 0:
            print("Incorrecto. ¡Sigue intentando!")
        else:
            print(f"Perdiste. El color era: {color_secreto}")
```
