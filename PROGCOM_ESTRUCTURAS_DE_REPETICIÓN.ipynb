#ESTRUCTURA DE REPETICIÓN
1. for. Cuando se sabe cuantas veces se debe repetir una acción
2. while. No se sabe cuantas veces se repite la acción.

##1. FOR
Sintaxis general en python
```
for (variable)
```
algo
* In range(n): quiero repetir n veces
* in range(CI, CF, p): Quiero repetir desde CI, hasta CF-1, con paso p
* in (lista, diccionario)
* for each (lista, diccionario): cuando quiero iterar cada elemento dentro de la lista/diccionario
*each (lista, diccionario): Cuando quiero iterar cada elemento dentro de la lista/diccionario



````python
#Quiero repetir n veces
for i in range(9):
  print(f"Hay {i+1} personas en el salon")
````

````python
for i in range(1,10):
  print(f"Hay {i} personas en el salon")
````

````python
dias=int(input("Ey!! Cuantos dias faltan para tu cumpleaños"))
for cumple in range(dias, 0, -1):
  print(f"Faltan {cumple} dias para tu cumpleaños")
  print("FELIZ CUMPLEAÑOS!")
````

#WHILE

Sintaxis
```
while (CI):
  lo que se va a repetir
  modificador de CI
```
**OJO!!**
- while arranca si CI es VERDADERA
- modificadores de CI: pregunta, contador, acumulador, cambio de variable


````python
c1=0
c2=0
c3=0
c4=0
nulo=0
voto=int(input("Ingrese el numero de su candidato (1-4). /n0 para voto en blanco"))
if voto==1: c1=c1+1
elif voto==2: c2+1 #c2+c2+1
elif voto==3: c3+=1
elif voto==4: c4+=1
else: nulo+=1
print("Votos para candidato 1:",c1)
print("Votos para candidato 2:",c2)
print("Votos para candidato 3:",c3)
print("Votos para candidato 4:",c4)
print("Votos para voto en blanco:",nulo)
````

````python
c1=0
c2=0
c3=0
c4=0
nulo=0
votos="s"
while votos=="s":
  voto=int(input("Ingrese el numero de su candidato (1-4). /n0 para voto en blanco"))
  if voto==1: c1=c1+1
  elif voto==2: c2+1 #c2+c2+1
  elif voto==3: c3+=1
  elif voto==4: c4+=1
  else: nulo+=1
  votos = input("Hay mas votos? s para si, n para no").lower()
  votos = votos[0]
  print("")
print("Votos para candidato 1:",c1)
print("Votos para candidato 2:",c2)
print("Votos para candidato 3:",c3)
print("Votos para candidato 4:",c4)
print("Votos para voto en blanco:",nulo)
````
